```python
import pandas as pd
import numpy as np

import os
import pathlib
import re

import matplotlib

```

```python
combinedData = pd.DataFrame()
files = [str(pathlib.Path("./original_data/") / f) for f in os.listdir("./original_data")]
for file in files:
    year = int(re.findall('\\d+', file)[0])
    df = pd.read_csv(file)
    df["year"] = year
    combinedData = pd.concat([combinedData, df], ignore_index=True)
```

```python
combinedData
```

```python
combinedData.drop(["College"], axis="columns", inplace=True)
```

```python
print(combinedData["Ht"])
```

```python


from numpy import isnan


def convert_height(height_string) -> int:
    feet, inches = height_string.split("-")
    feet, inches = int(feet), int(inches)
    total = 12 * feet + inches
    return total

combinedData["Ht"] = combinedData["Ht"].apply(lambda h: convert_height(h) if pd.notna(h) else h)
```

```python
combinedData.to_csv("./combine_2010_2025")
```
